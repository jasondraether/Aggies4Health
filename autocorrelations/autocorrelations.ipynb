```python
import pandas as pd
import interface as itf
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('new_data/day_data.csv', header=0)
```

```python
target_cols = [
    'laminutes',
    'maminutes',
    'vaminutes',
    'sedminutes',
    'resthr',
    'duration',
    'hr_mean',
    'steps_sum',
    'dist_sum',
    'cal_sum'
]
```

```python
for c in target_cols:
    x = range(1, 15, 1)
    autocors = [df[c].autocorr(i) for i in x]
    plt.plot(x, autocors)
    plt.title(c)
    plt.xlabel('Time Shift (Day)')
    plt.ylabel('Correlation')
    plt.axvline(x=7, color='red')
    #plt.axvline(x=30, color='green')
    plt.show()
    
```
